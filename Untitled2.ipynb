```python
#decisiones y ciclos (ejercicio 7)
x = (int)(input('digite un numero: '))
x = abs(x)
a = []
while (x >= 10):
    res = x%10
    a.append(res)
    x = x//10
a.append(x)#convertimos el numero a lista
a.reverse()#lo re_organizamos
print(a)
b = []#creamos una lista vacia
for i in range (0,len(a)):#iteramos dentro del rango de la lista a 
    for j in range(i+1,len(a)):#hacemos una iteracion dentro de la iteracion
        #                       para poder comparar los elementos de la lista
        if a[i] == a[j]:#si el elemtno de la primera iteracion es igual al 
            #            elemento de la segunda iteracion
            b.append(a[i])#sumamos este elemento a la lista vacia definida con 
            #              anterioridad
for i in range (0,len(b)):#hacemos otra iteracion dentro de una iteracion
    for j in range(i+1,len(b)):#para poder eliminar los elementos repetidos
        #                       en la liusta b
        if b[i] == b[j]:#al igual que como en la anterior iteracion comparamos
            #           los elementos de la lista
            b[j] = 'repit'#si hay un numero que se repite vamos a cambiar el 
            #             elemento por una cadena de texto
c = []#creamo una lista vacia
for i in range(0,len(b)):#iteramos en el rango de la lista b
        if b[i] != 'repit':#y los elementos que sean diferentes de 'repit'
            c.append(b[i])#van a ser sumados a la nueva lista vacia
print('los digitos que se repiten son: ',c)#de esta manera encontramos los
    #                                       digitos repetidos en el numero
    #                                       ingresado al inicio
```

```python
#decisiones y ciclos (ejercicio 9)
import math#importamos la libreria math
#definimos las entradas para las cordenadas y el radio de los dos circulos
x1=(int)(input('(primer circulo)introduzca la coordenada en x: '))
y1=(int)(input('(primer circulo)introduzca la coordenada en y: '))
r1=(int)(input('(primer circulo)introduzca el radio: '))

x2=(int)(input('(segundo circulo)introduzca la coordenada en x: '))
y2=(int)(input('(segundo circulo)introduzca la coordenada en y: '))
r2=(int)(input('(segundo circulo)introduzca el radio: '))
#definimos las coordenadas (a,b) de un punto
a=(int)(input('coord en a: '))
b=(int)(input('coord en b: '))
#aqui definimos la distancia del punto (a,b) a las cordenadasde los 2 circulos
dist1=math.sqrt((x1-a)**2+(y1-b)**2)
dist2=math.sqrt((x2-a)**2+(y2-b)**2)

cont=0
if (dist1<=r1):# si la distancia es menor al radio del circulo 1
    print('(a,b) esta en el circulo 1')#significa que el punto esta dentro
    #                                   del circulo 1
    cont=cont+1#iniciamos un contador para este caso
else:#si no se cumple la condicion anterior el punto no esta en el circulo 1
    print('(a,b) no esta en el circulo 1')
if (dist2<=r2):#si la distancia es menor al radio del circulo 2 
    print('(a,b) esta en el circulo 2')#significa que el punto esta dentro del
    #                                   circulo 2
    cont=cont+1#iniciamos un contador para este caso
else:#si no se cumple la condicio anterior el punto no esta en el circulo 2 
    #ninguno de los dos circulos
    print('(a,b) no esta en el circulo 2')
if (cont==2):#si nuestro contador es igual a 2 significa que cumplio las dos 
    #         condiciones anteriores por lo tanto el punto se encuentra en los
    #         2 circulos
    print('(a,b) esta en los dos circulos')
if (cont==0):#si el contador es 0 significa que que no se cumplio las dos 
    #         condiciones anteriores por lo tanto el punto no esta en ninguno
    #         de los dos circulos
    print('(a,b) no esta en ninguno de los dos circulos')
```

```python
#listas (ejercicio 18)
A_nxn = (int)(input('inserte el el tamaño n de la matriz cuadrada: '))#primero
#                              pedimos un numero  entero el cual nos dira el  
#                              tamaño de nuestra matriz cuadrada
A = []#creamos una lista vacia
for fila in range(A_nxn):#en el rango del numero entero
    A.append([0]*A_nxn)#multiplicamos una lista con un cero por el tamaño de la
    #                  matriz para asi crear una sublista con el nuemero de
    #                  elementos igual al numero de columnas de la matriz y esa
    #                  sublista la sumamos (A_nxn)veces a la lista vacia A para
    #                  crear una lista de listas, esto me va a representar una
    #                  matriz cuadrada de tamaño (A_nxn)
for fila in A:
    print(fila)#imprime la forma de la matriz pero la verdadera matriz es A, la
    #         matriz solo tiene cero porque el problema pide una matriz 
    #         triangular superior y estas matrices son matrices con 1 en la 
    #         diagonal de la matriz y en los elementos que quedan arriba de 
    #         siendo los demas elementos igual a 0
for i in range(0,A_nxn):#iteramos en las filas de la matriz
    for j in range(0,A_nxn):#hacemos una iteracion en las columnas de la matriz
        #                    sobre la iteracion en las filas de la maytriz
       if j >= i:#condicionamos para generar la matriz triangular
           A[i][j] = 1#de lo contrario toda la matriz se llenaria de unos
print('la matriz triangula asociada a A es: ', A)#imprime la matriz triangular   
f_max1 = max(A)#encontramos e imprimimos el elemento con mas unos que sera el 
#               elemento mas grande
print('el elemento 1 es la fila con mayor cantida de unos: ',f_max1)#se dice 
#que el elemento 1 es el elemento con mayor cantidad de unos porque una 
#matriz triangular superior la fila 1 va a ser la unica fila que haya unos en
#todas sus columnas
```

```python
#listas (ejercicio 19)
x = (int)(input('numero de elementos en la lista: '))#digitamos el numero de 
#                                        elementos que tendra la lista
lista = []#creamos una lista vacia
for i in range(0,x):#se itera desde 0 hasta el numero de elementos de la lista
    element = (float)(input('digite el elemento de la lista: '))#digitamos los
    #                                             elementos que tendra la lista
    lista.append(element)#esos elementos se suman a la lista vacia
print(lista)
sub_list = []#creamos otra lista vacia
for i in range(len(lista)-1):#iteramos en el rango de la lista menos 1 porque 
    #                     el ultimo valor no tendra una secuencia ascendente
    sub_list.append([])#a la sublista le agregamos otras sublistas por cada 
    #            elemento que tenga la lista inicial
    sub_list[i].append(lista[i])#a cada sublista le agregamos el valor 
    #                 correspondiente de la lista inicial
    for j in range(i+1,len(lista)):#iteramos en j(valor dentro de la sublista)
        #             con un rango desde el segundo elemento de la lista hasta
        #             el ultimo
        if lista[j-1]<=lista[j]:# si dentro de j aparece un valor que es menor
            # al valor que le sigue
            sub_list[i].append(lista[j])#en la sublista se le sumara el valor j
            
print('sub listas ordenadas ascendentes son:',sub_list)#imprimimos la lista con 
#                                      las sublistas de los valores ascendentes
tamaño = []#creamos una lista vacia
for i in range(len(sub_list)):#iteramos en el rango de la lista con las
    #              sublistas de valores ascendentes
    maxi = len(sub_list[i])#encontramos la longitud de cada lista
    tamaño.append(maxi)#a la lista vacia le sumamos los elementos que corres-
    #                   ponden a la longitud de la sublistas
print('el tamaño de las sub listas son: ',tamaño)
max_tamaño =max(tamaño)#encontramos el maximo elemento de la lista 
#                    correpondiente a la longitud de la sub listas este maximo
#                    pertenecera a la lista de mayor longitus

for i in range(len(sub_list)):
    if len(sub_list[i]) == max_tamaño:# si la longitud de la sublista es igual
        #                     al maximo encontrado anteriormente va a imprimir
        #                     la sublista que corresponde a ese maximo
        print('la lista ordenada ascendente de mayor tamaño es: ',sub_list[i])
```
